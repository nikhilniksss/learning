```python
c = np.array([2,-1,3])
d = np.array([1,4,-2])
```

```python
c + d
```

```python
# dot product

np.dot(c,d)
```

```python
# cross product

np.cross(c,d)
```

```python
# matrix

A = np.array([[1,2],[3,4]])
A
```

```python
B = np.array([[5,3],[3,4]])
B
```

```python
# matrix addition

A + B
```

```python
# matrix multiplication
 
A * B
```

```python
# Dot product

np.dot(A,B)
```

```python
# transpose of matrix by 180 degree
print(f"This is original matrix : {A}")
print(f" This is the transposed matrix : {A.T}")
```

```python
# Inverse of the matrix

np.linalg.inv(A)
```

```python
A
```

```python
# eigen vector

np.linalg.eig(A)
```
