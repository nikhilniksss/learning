```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.pipeline import make_pipeline
from sklearn.metrics import r2_score
```

```python
icecream_data = pd.read_csv(r"/Users/nikhilsingh/Desktop/FSDS/files/Icecream.csv")
icecream_data.head()
```

```python
icecream_data.columns
```

```python
plt.scatter(icecream_data['Temperature (°C)'],icecream_data['Ice Cream Sales (units)'])
```

```python
x = icecream_data[['Temperature (°C)']].values
y = icecream_data['Ice Cream Sales (units)'].values
```

```python
# polynomial model

poly_model = make_pipeline(PolynomialFeatures(degree=3),LinearRegression())
poly_model.fit(x,y)
y_pred = poly_model.predict(x)
```

```python
# plot to underdand which degree is important

plt.scatter(icecream_data['Temperature (°C)'],icecream_data['Ice Cream Sales (units)'])
plt.plot(x,y_pred,color = 'red')
```

```python
# try with other degree

# polynomial model - degree 4

poly_model_4 = make_pipeline(PolynomialFeatures(degree=4),LinearRegression())
poly_model_4.fit(x,y)
y_pred_4 = poly_model.predict(x)


plt.scatter(icecream_data['Temperature (°C)'],icecream_data['Ice Cream Sales (units)'])
plt.plot(x,y_pred_4,color = 'red')
```
